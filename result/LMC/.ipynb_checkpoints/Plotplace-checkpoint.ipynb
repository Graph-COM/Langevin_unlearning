# Fig 1, vs Neal et al.

```python
FS = 20
LFS = 10

eps = np.array([0.05,0.1,0.5,1.0,2.0,5.0])
D2D_dict = {}
LU_dict = {}
RT_dict = {}
K_list = np.array([1,2,5])

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,6), sharex=True)

for dataset in ['MNIST','CIFAR10']:
    for K in [1,2,5]:
        D2DResults = [[],[]]
        LUResults = [[],[]]
        RTResults = [[],[]]
        for e in [0.05,0.1,0.5,1,2,5]:
            D2Dpath = f'./{dataset}/baseline/{K}/baseline_acc_unlearn_finetune{e}.npy'
            LUpath = f'./{dataset}/baseline/{K}/lmc_acc_unlearn_finetune{e}.npy'
            RTpath = f'./{dataset}/baseline/{K}/lmc_acc_unlearn_scratch{e}.npy'

            Result = np.load(D2Dpath)
            Mean, Std = Result.mean(), Result.std()
            D2DResults[0].append(Mean)
            D2DResults[1].append(Std)

            Result = np.load(LUpath)
            Mean, Std = Result.mean(), Result.std()
            LUResults[0].append(Mean)
            LUResults[1].append(Std)
            
            Result = np.load(RTpath)
            Mean, Std = Result.mean(), Result.std()
            RTResults[0].append(Mean)
            RTResults[1].append(Std)

        D2D_dict[f'{dataset}_{K}'] = np.array(D2DResults)
        LU_dict[f'{dataset}_{K}'] = np.array(LUResults)
        RT_dict[f'{dataset}_{K}'] = np.array(RTResults)
            

d_counter = 0
for dataset in ['MNIST','CIFAR10']:
    counter = 0
    for K in [1,2,5]:
        if K in [1]:
            ax[d_counter].semilogx(eps,LU_dict[f'{dataset}_{K}'][0],marker='o',label=f'Langevin Unlearning')
            ax[d_counter].fill_between(eps, LU_dict[f'{dataset}_{K}'][0]-LU_dict[f'{dataset}_{K}'][1], LU_dict[f'{dataset}_{K}'][0]+LU_dict[f'{dataset}_{K}'][1], color='C0', alpha=0.3)
            
            ax[d_counter].semilogx(eps,RT_dict[f'{dataset}_{K}'][0],marker='o',label=f'Langevin Retraining')
            ax[d_counter].fill_between(eps, RT_dict[f'{dataset}_{K}'][0]-RT_dict[f'{dataset}_{K}'][1], RT_dict[f'{dataset}_{K}'][0]+RT_dict[f'{dataset}_{K}'][1], color='C1', alpha=0.3)
        
        ax[d_counter].semilogx(eps,D2D_dict[f'{dataset}_{K}'][0],marker='o', color=f'C{2+counter}',label=f'D2D, $K=${K}')
        ax[d_counter].fill_between(eps, D2D_dict[f'{dataset}_{K}'][0]-D2D_dict[f'{dataset}_{K}'][1], D2D_dict[f'{dataset}_{K}'][0]+D2D_dict[f'{dataset}_{K}'][1], color=f'C{2+counter}', alpha=0.3)
    
        # ax[d_counter].set_title(f'{dataset}', size=FS)
        # ax[d_counter,counter].set_ylabel('Accuracy', fontsize=FS)
        # ax[d_counter,counter].set_xlabel('$\epsilon$', fontsize=FS)
        # ax[d_counter,counter].label_outer()
        counter = counter + 1
    d_counter = d_counter + 1

ax[0].legend(shadow=True, loc='lower right',fontsize=LFS)
fig.supylabel('Accuracy', fontsize=FS)
fig.supxlabel('$\epsilon$', fontsize=FS)
plt.tight_layout()
plt.show()
```

# Fig 2-1, our own $\epsilon$ vs $K$ for different batch unlearning size $S$

```python
FS = 20
LFS = 10

eps = np.array([0.1,0.5,1.0,2.0,5.0])
LU_dict = {}
S_list = np.array([10,300,1000])
K_dict = {}
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(4,6), sharex=True, sharey=True,layout='constrained')

for dataset in ['MNIST','CIFAR10']:
    temp = np.load(f'./{dataset}/paint_utility_epsilon/K_list.npy', allow_pickle = True)
    temp = eval(str(temp).split("array(")[0])
    for S in [10,300,1000]:
        K_list = []
        LUResults = [[],[]]
        for e in [0.1,0.5,1,2,5]:
            K_list.append(temp[S][e])
            
            LUpath = f'./{dataset}/paint_utility_epsilon/{S}/acc_finetune_epsilon{e}.npy'

            Result = np.load(LUpath)
            Mean, Std = Result.mean(), Result.std()
            LUResults[0].append(Mean)
            LUResults[1].append(Std)

        LU_dict[f'{dataset}_{S}'] = np.array(LUResults)
        K_dict[S] = K_list    
d_counter = 0
for dataset in ['MNIST','CIFAR10']:
    counter = 0
    for S in [10,300,1000]:
        acc_mean, acc_std = np.round(LU_dict[f'{dataset}_{S}'][0].mean(),3), LU_dict[f'{dataset}_{S}'][1].std()
        ax[d_counter].plot(eps[1:],K_dict[S][1:],marker='o',label=f'$S=${S}, acc={str(acc_mean)}')
        
        # for i, txt in enumerate(LU_dict[f'{dataset}_{S}'][0]):
        #     ax[d_counter].annotate(np.round(txt,3), (eps[i], K_dict[S][i]))
        
        # ax[d_counter].set_title(f'{dataset}', size=FS)
        ax[d_counter].legend(shadow=True, loc='upper right',fontsize=LFS)
        counter = counter + 1
    d_counter = d_counter + 1


fig.supylabel('Unlearning iteration $K$', fontsize=FS)
fig.supxlabel('           $\epsilon$', fontsize=FS)


# plt.tight_layout()
plt.show()
```

# Fig 2-2, our own $S$ vs Acc for different $\epsilon$, compared with retraining

```python
FS = 20
LFS = 10

eps = np.array([0.5,1.0,2.0])
LU_dict = {}
RT_dict = {}
S_list = np.array([10, 50, 100, 500, 1000])
K_dict = {}
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5,6), sharex=True,layout='constrained')


for dataset in ['MNIST','CIFAR10']:
    learn_scratch_list = np.load(f'./{dataset}/paint_utility_s/acc_scratch_D.npy', allow_pickle = True)
    for e in [0.5,1,2]:
        K_list = np.load(f'./{dataset}/paint_utility_s/{e}/K_list.npy', allow_pickle = True)
        LUResults = [[],[]]
        RTResults = [[],[]]
        for S in S_list:
            LUpath = f'./{dataset}/paint_utility_s/{e}/acc_finetune_remove{S}.npy'
            RTpath = f'./{dataset}/paint_utility_s/{e}/acc_scratch_Dnew_remove{S}.npy'

            Accs = np.load(LUpath)
            LUResults[0].append(Accs.mean())
            LUResults[1].append(Accs.std())

            Accs = np.load(RTpath)
            RTResults[0].append(Accs.mean())
            RTResults[1].append(Accs.std())

        LU_dict[f'{dataset}_{e}'] = np.array(LUResults)
        RT_dict[f'{dataset}_{e}'] = np.array(RTResults)
        K_dict[f'{dataset}_{e}'] = K_list

        
d_counter = 0
for dataset in ['MNIST','CIFAR10']:
    counter = 0
    for e in [0.5,1,2]:
        ax[d_counter].plot(S_list,LU_dict[f'{dataset}_{e}'][0],marker='o', color=f'C{counter}',label=f' Unlearning, $\epsilon=${e}')
        ax[d_counter].fill_between(S_list,LU_dict[f'{dataset}_{e}'][0]-LU_dict[f'{dataset}_{e}'][1], LU_dict[f'{dataset}_{e}'][0]+LU_dict[f'{dataset}_{e}'][1], color=f'C{counter}', alpha=0.1)
        ax[d_counter].plot(S_list,RT_dict[f'{dataset}_{e}'][0],marker='o', color=f'C{counter}',label=f' Retraining, $\epsilon=${e}', linestyle = '--')
        ax[d_counter].fill_between(S_list,RT_dict[f'{dataset}_{e}'][0]-RT_dict[f'{dataset}_{e}'][1], RT_dict[f'{dataset}_{e}'][0]+RT_dict[f'{dataset}_{e}'][1], color=f'C{counter}', alpha=0.1)
        # for i, txt in enumerate(K_dict[f'{dataset}_{e}']):
        #     if i >= 5:
        #         continue
        #     ax[d_counter].annotate(f'K={txt}', (S_list[i], LU_dict[f'{dataset}_{e}'][0][i]))
        
        # ax[d_counter].set_title(f'{dataset}', size=FS)

        counter = counter + 1
    d_counter = d_counter + 1

ax[1].legend(shadow=True, loc='lower left',fontsize=LFS,ncol=2)
fig.supylabel('Accuracy', fontsize=FS)
fig.supxlabel('             $S$', fontsize=FS)
# plt.tight_layout()
plt.show()
```

# Fig 3, $\sigma$ vs $\epsilon_0$ and accuracy. Compare with retraining and initial learning

```python
FS = 20
LFS = 10

sigma_list = [0.05, 0.1, 0.2, 0.5, 1]

LL_dict = {}
LU_dict = {}
RT_dict = {}
epsilon0_dict = {}
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,6), sharex=True,layout='constrained')


for dataset in ['MNIST','CIFAR10']:
    LUResults = [[],[]]
    LLResults = [[],[]]
    RTResults = [[],[]]
    for sigma in sigma_list:
        LUpath = f'./{dataset}/paint_unlearning_sigma/{str(sigma)}_acc_finetune.npy'
        LLpath = f'./{dataset}/paint_unlearning_sigma/{str(sigma)}_acc_scratch_D.npy'
        RTpath = f'./{dataset}/paint_unlearning_sigma/{str(sigma)}_acc_scratch_Dnew.npy'

        Accs = np.load(LUpath)
        LUResults[0].append(Accs.mean())
        LUResults[1].append(Accs.std())

        Accs = np.load(LLpath)
        LLResults[0].append(Accs.mean())
        LLResults[1].append(Accs.std())
        
        Accs = np.load(RTpath)
        RTResults[0].append(Accs.mean())
        RTResults[1].append(Accs.std())

    LU_dict[f'{dataset}'] = np.array(LUResults)
    LL_dict[f'{dataset}'] = np.array(LLResults)
    RT_dict[f'{dataset}'] = np.array(RTResults)
    epsilon0_dict[f'{dataset}'] = np.load(f'./{dataset}/paint_unlearning_sigma/epsilon0.npy')

# Save needed K
K_dict = {}
for dataset in ['MNIST','CIFAR10']:
    for sigma in sigma_list:
        temp = np.load(f'./{dataset}/paint_unlearning_sigma/K_dict{sigma}.npy',allow_pickle=True)
        temp = eval(str(temp).split("array(")[0])
        K_dict[f'{dataset}_{sigma}'] = temp[1000][1]

d_counter = 0
for dataset in ['MNIST','CIFAR10']:
    ax[d_counter].semilogy(sigma_list, epsilon0_dict[f'{dataset}'], label = r'$\epsilon_0$', marker='^', linestyle='--', color='k')
    ax[d_counter].set_ylabel(r'$\epsilon_0$', fontsize = FS)  
    ax[d_counter].tick_params(axis='y')  
    ax2 = ax[d_counter].twinx()
    ax2.plot(sigma_list, LU_dict[f'{dataset}'][0], marker='o', color='C0',label=f' Unlearning')

    rt_acc_list = []
    for sigma in sigma_list:
        tmp_rt = np.load(f'./{dataset}/retrain/{sigma}_acc_scratch_D.npy', allow_pickle = True)
        rt_acc_list.append(np.mean(tmp_rt))
    ax2.plot(sigma_list, rt_acc_list, label = 'retrain K')

    for i, sigma in enumerate(sigma_list):
        txt = K_dict[f'{dataset}_{sigma}']
        if i > 3:
            ax2.annotate(f'K={txt}', (sigma_list[i], LU_dict[f'{dataset}'][0][i]), ha='right')
        else:
            ax2.annotate(f'K={txt}', (sigma_list[i], LU_dict[f'{dataset}'][0][i]))

    ax2.plot(sigma_list, LL_dict[f'{dataset}'][0], marker='o', color='C1',label=f' Learning')
    ax2.plot(sigma_list, RT_dict[f'{dataset}'][0], marker='o', color='C2',label=f' Retraining')

    ax2.fill_between(sigma_list, LU_dict[f'{dataset}'][0]-LU_dict[f'{dataset}'][1], LU_dict[f'{dataset}'][0]+LU_dict[f'{dataset}'][1], color='C0', alpha=0.1)
    ax2.fill_between(sigma_list, LL_dict[f'{dataset}'][0]-LL_dict[f'{dataset}'][1], LL_dict[f'{dataset}'][0]+LL_dict[f'{dataset}'][1], color='C1', alpha=0.1)
    ax2.fill_between(sigma_list, RT_dict[f'{dataset}'][0]-RT_dict[f'{dataset}'][1], RT_dict[f'{dataset}'][0]+RT_dict[f'{dataset}'][1], color='C2', alpha=0.1)

    ax2.set_ylabel('Accuracy', fontsize = FS)
    ax2.tick_params(axis='y')
    d_counter = d_counter + 1

handles1, labels1 = ax[1].get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

handles = handles1 + handles2
labels = labels1 + labels2

### This prevents the legend being overlapped with the white line.
legend_1 = ax[1].legend(handles, labels, shadow=True, loc='upper right',fontsize=LFS)
legend_1.remove()

all_axes = fig.get_axes()
for axis in all_axes:
    legend = axis.get_legend()
    if legend is not None:
        legend.remove()
all_axes[-1].add_artist(legend_1)
###

fig.supxlabel('Noise standard deviation $\sigma$', fontsize=FS)
# plt.tight_layout()
plt.show()
```

```python
# Unlearn 100 points, eps = 1

FS = 20
LFS = 10

Max_nr = 100
D2D_dict = {}
LU_dict = {}
K_LU_dict = {}
K_D2D_list = np.array([1,5,10])
S_LU_list = np.array([5,10,20])


fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,6), sharex=True,layout='constrained')

for dataset in ['MNIST','CIFAR10']:
    # Baseline results
    for K in K_D2D_list:
        Max_step = int(Max_nr/K)
        D2DResults = [[],[]]
        for step in range(Max_step):
            D2Dpath = f'./{dataset}/sequential/baseline_acc_k_{K}_step_{step}.npy'

            Result = np.load(D2Dpath)
            Mean, Std = Result.mean(), Result.std()
            D2DResults[0].append(Mean)
            D2DResults[1].append(Std)
            
        D2D_dict[f'{dataset}_{K}'] = np.array(D2DResults)

    # Our results
    for S in S_LU_list:
        Max_step = int(Max_nr/S)
        LUResults = [[],[]]
        for step in range(Max_step):
            LUpath = f'./{dataset}/sequential/lmc_acc_finetune_nr{S}_step{step}.npy'

            Result = np.load(LUpath)
            Mean, Std = Result.mean(), Result.std()
            LUResults[0].append(Mean)
            LUResults[1].append(Std)
            
        LU_dict[f'{dataset}_{S}'] = np.array(LUResults)

        # Corresponding K
        K_LU_dict[f'{dataset}_{S}'] = np.load(f'./{dataset}/sequential/k_list_nr{S}.npy')
            

d_counter = 0
for dataset in ['MNIST','CIFAR10']:
    counter = 0
    for S in S_LU_list:
        Max_step = int(Max_nr/S)
        # number of current nr.
        x = np.linspace(0,Max_nr,Max_step+1)
        x = x[1:]
        # ax[d_counter].plot(x,LU_dict[f'{dataset}_{S}'][0],marker='o', color=f'C{counter}',label=f'Langevin Unlearning, $S=${S}')
        # ax[d_counter].fill_between(x, LU_dict[f'{dataset}_{S}'][0]-LU_dict[f'{dataset}_{S}'][1], LU_dict[f'{dataset}_{S}'][0]+LU_dict[f'{dataset}_{S}'][1], color=f'C{counter}', alpha=0.3)
        ax[d_counter].plot(x,np.cumsum(K_LU_dict[f'{dataset}_{S}'][1:]),marker='o', color=f'C{counter}',label=f'Langevin Unlearning, $S=${S}')

        # Annotate start and end acc
        # Acc = np.round(LU_dict[f'{dataset}_{S}'][0][0],3)
        # ax[d_counter].annotate(f'{str(Acc)}', (x[0], np.cumsum(K_LU_dict[f'{dataset}_{S}'][1:])[0]))
        Acc = np.round(LU_dict[f'{dataset}_{S}'][0][-1],3)
        ax[d_counter].annotate(f'{str(Acc)}', (x[-1], np.cumsum(K_LU_dict[f'{dataset}_{S}'][1:])[-1]), ha='right')
        # ax[d_counter,counter].semilogx(eps,RT_dict[f'{dataset}_{K}'][0],marker='o',label=f'Langevin Retraining')
        # ax[d_counter,counter].fill_between(eps, RT_dict[f'{dataset}_{K}'][0]-RT_dict[f'{dataset}_{K}'][1], RT_dict[f'{dataset}_{K}'][0]+RT_dict[f'{dataset}_{K}'][1], color='C1', alpha=0.3)
        
        counter = counter + 1

    for K in K_D2D_list:
        Max_step = Max_nr
        x = np.linspace(0,Max_nr,Max_step+1)
        x = x[1:]
        ax[d_counter].plot(x,x*K, color=f'C{counter}',label=f'D2D, $K=${K}')
        counter = counter + 1
    d_counter = d_counter + 1

ax[0].legend(shadow=True, loc='upper left',fontsize=LFS)
# fig.supylabel('Accuracy', fontsize=FS)
fig.supylabel('Accumulate K', fontsize=FS)
fig.supxlabel('         Number of unlearned data', fontsize=FS)
# plt.tight_layout()
plt.show()
```
