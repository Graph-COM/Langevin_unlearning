```python
import matplotlib.pyplot as plt
import seaborn as sns

from adjustText import adjust_text

import numpy as np
sns.set_theme()

plt.rcParams["mathtext.fontset"] = "cm"
```

# retrain noiseless MNIST and CIFAR10

```python
FS = 20
LFS = 10

eps = np.array([0.05,0.1,0.5,1.0,2.0,5.0])
D2D_dict = {}
LU_dict = {}
RT_dict = {}
noiseless_dict = {}

K_list = np.array([1,2,5])

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,6), sharex=True)

for dataset in ['MNIST','CIFAR10']:
    for K in [1,2,5]:
        D2DResults = [[],[]]
        LUResults = [[],[]]
        RTResults = [[],[]]
        noiselessResults = [[], []]
        for e in [0.05,0.1,0.5,1,2,5]:
            D2Dpath = f'./{dataset}/baseline/{K}/baseline_acc_unlearn_finetune{e}.npy'
            LUpath = f'./{dataset}/baseline/{K}/lmc_acc_unlearn_finetune{e}.npy'
            RTpath = f'./{dataset}/baseline/{K}/lmc_acc_unlearn_scratch{e}.npy'
            noiselesspath = f'./{dataset}/retrain_noiseless/retrain_noiseless1.npy'

            Result = np.load(D2Dpath)
            Mean, Std = Result.mean(), Result.std()
            D2DResults[0].append(Mean)
            D2DResults[1].append(Std)

            Result = np.load(LUpath)
            Mean, Std = Result.mean(), Result.std()
            LUResults[0].append(Mean)
            LUResults[1].append(Std)
            
            Result = np.load(RTpath)
            Mean, Std = Result.mean(), Result.std()
            RTResults[0].append(Mean)
            RTResults[1].append(Std)

            Result = np.load(noiselesspath)
            Mean, Std = Result.mean(), Result.std()
            noiselessResults[0].append(Mean)
            noiselessResults[1].append(Std)
        
        
        D2D_dict[f'{dataset}_{K}'] = np.array(D2DResults)
        LU_dict[f'{dataset}_{K}'] = np.array(LUResults)
        RT_dict[f'{dataset}_{K}'] = np.array(RTResults)
        noiseless_dict[f'{dataset}_{K}'] = np.array(noiselessResults)
            

d_counter = 0
for dataset in ['MNIST','CIFAR10']:
    counter = 0
    for K in [1,2,5]:
        if K in [1]:
            ax[d_counter].semilogx(eps,noiseless_dict[f'{dataset}_{K}'][0],marker='o',label=f'Langevin noiseless')
            ax[d_counter].fill_between(eps, noiseless_dict[f'{dataset}_{K}'][0]-noiseless_dict[f'{dataset}_{K}'][1], noiseless_dict[f'{dataset}_{K}'][0]+noiseless_dict[f'{dataset}_{K}'][1], color='C100', alpha=0.3)
            
            ax[d_counter].semilogx(eps,LU_dict[f'{dataset}_{K}'][0],marker='o',label=f'Langevin Unlearning')
            ax[d_counter].fill_between(eps, LU_dict[f'{dataset}_{K}'][0]-LU_dict[f'{dataset}_{K}'][1], LU_dict[f'{dataset}_{K}'][0]+LU_dict[f'{dataset}_{K}'][1], color='C0', alpha=0.3)
            
            ax[d_counter].semilogx(eps,RT_dict[f'{dataset}_{K}'][0],marker='o',label=f'Langevin Retraining')
            ax[d_counter].fill_between(eps, RT_dict[f'{dataset}_{K}'][0]-RT_dict[f'{dataset}_{K}'][1], RT_dict[f'{dataset}_{K}'][0]+RT_dict[f'{dataset}_{K}'][1], color='C1', alpha=0.3)
        
        ax[d_counter].semilogx(eps,D2D_dict[f'{dataset}_{K}'][0],marker='o', color=f'C{2+counter}',label=f'D2D, $K=${K}')
        ax[d_counter].fill_between(eps, D2D_dict[f'{dataset}_{K}'][0]-D2D_dict[f'{dataset}_{K}'][1], D2D_dict[f'{dataset}_{K}'][0]+D2D_dict[f'{dataset}_{K}'][1], color=f'C{2+counter}', alpha=0.3)
    
        # ax[d_counter].set_title(f'{dataset}', size=FS)
        # ax[d_counter,counter].set_ylabel('Accuracy', fontsize=FS)
        # ax[d_counter,counter].set_xlabel('$\epsilon$', fontsize=FS)
        # ax[d_counter,counter].label_outer()
        counter = counter + 1
    d_counter = d_counter + 1

ax[0].legend(shadow=True, loc='lower right',fontsize=LFS)
fig.supylabel('Accuracy', fontsize=FS)
fig.supxlabel('$\epsilon$', fontsize=FS)
plt.tight_layout()
plt.show()
```
